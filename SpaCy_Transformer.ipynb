```python
import json

def convert_conll_to_spacy(conll_file, spacy_file):
    converted_data = []
    with open(conll_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            if line == '' or line.startswith('-DOCSTART-'):
                continue
            tokens = line.split()
            token = tokens[0]
            ner_label = tokens[-1]
            if ner_label != 'O':
                start = line.index(token)
                end = start + len(token)
                entity = (start, end, ner_label)
                converted_data[-1]['entities'].append(entity)
            else:
                converted_data.append({'text': token, 'entities': []})

    with open(spacy_file, 'w') as f:
        for data in converted_data:
            f.write(json.dumps(data))
            f.write('\n')
conell_path = "Downloads/FIRST_IOB_ANNOTATED_DATA.conll"
convert_conll_to_spacy(conell_path, 'ner_data_spacy.json')
```
